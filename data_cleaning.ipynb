# Erste Erkenntnisse

## Unntige Daten
Um nhere Infos ber die Spalten zu erhalten, haben wir uns die Details und Anzahl der Unqiue-Werte pro Spalte ermittelt.
Folgende Spalten haben lediglich einen Uniquen-Wert:
- Fwd URG Flags 
- Bwd URG Flags 
- Fwd Byts/b Avg
- Fwd Pkts/b Avg
- Bwd Byts/b Avg
- Bwd Pkts/b Avg
- Fwd Blk Rate Avg
- Bwd Blk Rate Avg

Da diese Attribute lediglich einen Wert annehmen knnen, sind diese nicht hilfreich bei der Unterscheidung zwischen gutartig & DDoS. Im folgenden werden diese Werte daher aus dem Datensatz entfernt --> 10% der ursprnglichen Spalten entfernt.

## Null-Values
- Flow Byts/s
Die Spalte Flow Byts/s hat als einzige Spalte 29688 Nullwerte (nicht definiert NaN), was wir durch .info erkannt haben. Es gibt zwei Mglichkeiten, damit umzugehen.
- Option 1: Droppen
  - Ein Droppen wre sinnvoll, da wir dadurch keine falschen Annahmen ber die Daten treffen. Zudem machen diese Daten gerade erst einmal 0,4 % der Daten aus. 
- Option 2: Werte ersetzen 
  - Das Ersetzen der Werte wrde die Timeline (timestamps) nicht knstlich verndern (evtl. problematisch bei manchen ML-Modellen).

Vorest entscheiden wir uns fr Option 1.

## Infinity-Values
Bei der Untersuchung ist uns aufgefallen, dass Flow Byts/s und Flow Pkts/s in 0,2% der Daten Infinity-Werte aufweisen. Bei diesen Daten handelt es sich gutartige (Benign) Anfragen handelt. Deshalb haben wir uns dazu entschieden, diese Daten zu entfernen.

```python
df_inf = df[~da.isinf(df["Flow Byts/s"])]
df_nonull = df_inf.dropna(subset=["Flow Byts/s"])
df_clean = df_nonull.drop(["Fwd URG Flags", "Bwd URG Flags", "Fwd Byts/b Avg", "Fwd Pkts/b Avg", "Bwd Byts/b Avg", "Bwd Pkts/b Avg", "Fwd Blk Rate Avg", "Bwd Blk Rate Avg"], axis=1)
df_clean.to_csv('cleaned_ds.csv', single_file = True, index = False)
```
