## Insights Gained

### Not Informative Data
To gain further insights about the columns, we have determined the details and amount of unique values for column.
The following columns contain only one unique value:
- Fwd URG Flags 
- Bwd URG Flags 
- Fwd Byts/b Avg
- Fwd Pkts/b Avg
- Bwd Byts/b Avg
- Bwd Pkts/b Avg
- Fwd Blk Rate Avg
- Bwd Blk Rate Avg

Since these attributes can only take on one value, we cannot gain any information about the distinction between benign and ddos. Therefore, these columns will be removed. That means that we drop about 10% of the origin columns.

### Null-Values

The columns 'Flow Byts/s' is the only columns containing 29688 null values (NaN - Not a Number), which we discovered through the use of .info. There are to possible options for handling this:
- Option 1: Drop
  - Dropping the values seems sensible since we do not make any wrong assumptions about the data. Furthermore, the columns only contain 0.4% of the overall data.
- Option 2: Replace
  - Replacing the values does not lead to artifical changes within the timeline (timestamps) which might cause troubles for some models.

For now, we decided to use option 1.

## Infinity-Values
During the exploration of the data, we have noticed infinity values within 0.2% of the data of the columns 'Flow Byts/s' and 'Flow Pkts/s'. All of the rows with these values are labled 'benign'. For this reason, we decided to drop the rows as well.

---
## Cleaning

```python
df_inf = df[~da.isinf(df['Flow Byts/s'])]
df_nonull = df_inf.dropna(subset=['Flow Byts/s'])
df_clean = df_nonull.drop(['Fwd URG Flags', 'Bwd URG Flags', 'Fwd Byts/b Avg', 'Fwd Pkts/b Avg', 'Bwd Byts/b Avg', 'Bwd Pkts/b Avg', 'Fwd Blk Rate Avg', 'Bwd Blk Rate Avg'], axis=1)
df_clean.to_csv('cleaned_ds.csv', single_file = True, index = False)
```
